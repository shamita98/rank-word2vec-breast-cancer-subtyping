### Overview

Extraction of survival data of SCAN-B HiSeq samples for survival analysis

```python
import pandas as pd
import numpy as np
```

### 1. Extract survival data of SCAN-B HiSeq samples

```python
# import clinical metadata of scan-b hiseq samples
metadata_hiseq = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/datasets/SCANB_GSE202203/scanb_uncleaned_data/GSE202203_GPL11154_HiSeq_series_matrix_metadata.txt", skiprows=38, sep='\t').transpose()
```

```python
metadata_hiseq.head(3)
```

```python
# make row 0 as the column name
metadata_hiseq_copy = metadata_hiseq.copy()
metadata_hiseq_copy.columns = metadata_hiseq_copy.iloc[0]

# exclude row 0 (duplicate of the column name)
metadata_hiseq_copy = metadata_hiseq_copy.iloc[1:,:]
metadata_hiseq_copy.head(3)
```

```python
# extract patient characteristics columns
hiseq_patient_char = metadata_hiseq_copy.loc[:, ['!Sample_characteristics_ch1']]
hiseq_patient_char.head(3)
```

```python
# extract survival days, years and events that are in columns 15, 16 and 17
hiseq_survival = hiseq_patient_char.iloc[:,[15, 16, 17]]
hiseq_survival.head(3)
```

```python
# rename column names
hiseq_survival.columns = ['overall_survival_days', 'overall_survival_years', 'overall_survival_event']
hiseq_survival.head(3)
```

### 2. Convert survival data into suitable format

```python
# function to keep only numbers in the survival data
def keep_numbers_in_survival(df, col_names):
    df_copy = df.copy()
    df_nrow = df_copy.shape[0]
    for col in col_names:
        for i in range(df_nrow):
            value = df_copy.loc[df_copy.index[i], col]
            df_copy.loc[df_copy.index[i], col] = value.split(': ')[1]

    return df_copy
```

```python
# keep only the numeric data
hiseq_survival_modified = keep_numbers_in_survival(hiseq_survival, hiseq_survival.columns.tolist())
hiseq_survival_modified.head(3)
```

```python
# convert data types
hiseq_survival_modified['overall_survival_event'] = pd.to_numeric(hiseq_survival_modified['overall_survival_event'], errors ='coerce').astype('Int64')
hiseq_survival_modified['overall_survival_days'] = pd.to_numeric(hiseq_survival_modified['overall_survival_days'], errors ='coerce').astype('Int64')
hiseq_survival_modified['overall_survival_years'] = pd.to_numeric(hiseq_survival_modified['overall_survival_years'], errors ='coerce').astype('Float64')
```

```python
hiseq_survival_modified.head(3)
```

```python
# check data types
hiseq_survival_modified.dtypes
```

```python
# check the proportion of survival events
hiseq_survival_modified['overall_survival_event'].value_counts()
```

```python
# check for missing values
hiseq_survival_modified.isna().sum(axis=0)
```

```python
# check which samples (rows) have missing values
hiseq_survival_modified.loc[hiseq_survival_modified.isna().any(axis=1),:]
```

```python
# remove the sample with missing survival data
hiseq_survival_modified = hiseq_survival_modified.loc[~hiseq_survival_modified.isna().any(axis=1),:]
hiseq_survival_modified.shape
```

```python
# round the overall_survival_years to 3 decimal places
hiseq_survival_modified.loc[:,'overall_survival_years'] = round(hiseq_survival_modified['overall_survival_years'],3)
hiseq_survival_modified.head(3)
```

```python
# check the maximum overall survival years in the whole hiseq dataset (including unclassified and normal-like samples)
hiseq_survival_modified.loc[hiseq_survival_modified['overall_survival_years'].idxmax(),'overall_survival_years']
```

### 3. Split survival data according to SCAN-B HiSeq training and test sets

```python
# import scanb hiseq training and test data
scanb_hiseq_train = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/datasets/SCANB_GSE202203/scanb_hiseq_train_test_sets/train_test_80_20/SCANB_HiSeq_pam50gene_raw_counts_subtype_train_80.csv", 
                          header=0, index_col=0)

scanb_hiseq_test = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/datasets/SCANB_GSE202203/scanb_hiseq_train_test_sets/train_test_80_20/SCANB_HiSeq_pam50gene_raw_counts_subtype_test_20.csv", 
                          header=0, index_col=0)
```

```python
# select only the rows from gene expression_subtype dataset that are in survival dataset
# reorder the rows in the gene expression_subtype dataset according to the survival dataset
# note: the sample with missing survival data is in the training dataset, and is removed
scanb_hiseq_train_sorted = scanb_hiseq_train.loc[hiseq_survival_train.index,:]
scanb_hiseq_test_sorted = scanb_hiseq_test.loc[hiseq_survival_test.index,:]
print(scanb_hiseq_train_sorted.shape)
print(scanb_hiseq_test_sorted.shape)

# check whether the indices match
print(scanb_hiseq_train_sorted.index.equals(hiseq_survival_train.index))
print(scanb_hiseq_test_sorted.index.equals(hiseq_survival_test.index))
```

### 4. Merge survival data with subtype labels

```python
# merge survival and subtype data
hiseq_train_survival_subtype = hiseq_survival_train.join(scanb_hiseq_train_sorted.loc[:,'subtype'])
hiseq_test_survival_subtype = hiseq_survival_test.join(scanb_hiseq_test_sorted.loc[:,'subtype'])
print(hiseq_train_survival_subtype.shape)
print(hiseq_test_survival_subtype.shape)
```
