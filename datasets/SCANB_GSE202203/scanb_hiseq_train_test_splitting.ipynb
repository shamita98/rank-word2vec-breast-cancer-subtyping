### Overview

Splitting of SCAN-B HiSeq raw and TPM-normalized counts dataset into 80% training set and 20% test set.

Note: These datasets are with PAM50 genes as features.

```python
import pandas as pd
import numpy as np
import random
from sklearn.model_selection import train_test_split
```

### 1. Import and Prepare SCANB Hiseq PAM50 Raw and TPM Counts Data

```python
# import scanb hiseq raw counts of PAM50 genes
scanb_raw_data = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/datasets/SCANB_GSE202203/scanb_cleaned_data/pam50genes_raw_counts_subtype_scanb_hiseq.csv",
                       header=0, index_col=0)
```

```python
# import scanb hiseq tpm counts of PAM50 genes
scanb_tpm_data = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/datasets/SCANB_GSE202203/scanb_cleaned_data/pam50genes_tpm_counts_subtype_scanb_hiseq.csv",
                       header=0, index_col=0)
```

```python
# check whether the indices and columns match
print(scanb_raw_data.index.equals(scanb_tpm_data.index))
print(scanb_raw_data.columns.equals(scanb_tpm_data.columns))

# check whether the subtype order matches
print(scanb_raw_data['subtype'].tolist() == scanb_tpm_data['subtype'].tolist())
```

```python
print(scanb_raw_data.shape)
print(scanb_tpm_data.shape)
```

```python
# check subtype proportion in raw and tpm datasets
print('Raw dataset:')
print(scanb_raw_data['subtype'].value_counts())

print('\nTPM dataset:')
print(scanb_tpm_data['subtype'].value_counts())
```

```python
# separate counts and labels
scanb_raw_counts = scanb_raw_data.iloc[:, 0:50]
scanb_tpm_counts = scanb_tpm_data.iloc[:, 0:50]
scanb_subtype = scanb_raw_data.iloc[:, [50]]
```

### 2. Split the SCAN-B Hiseq Dataset into 80% Training and 20% Test

```python
# note: initialize randomness before splitting
seed = 42
random.seed(seed)
np.random.seed(seed)

# split scanb raw and tpm data into 80% training and 20% test sets
scanb_raw_train, scanb_raw_test, scanb_raw_subtype_train, scanb_raw_subtype_test = train_test_split(scanb_raw_counts, scanb_subtype, test_size=0.20, 
                                                                                            stratify=scanb_subtype, shuffle=True, random_state=seed)

scanb_tpm_train, scanb_tpm_test, scanb_tpm_subtype_train, scanb_tpm_subtype_test = train_test_split(scanb_tpm_counts, scanb_subtype, test_size=0.20,
                                                                                            stratify=scanb_subtype, shuffle=True, random_state=seed)
```

```python
# check whether the indices match between the raw and tpm datasets
print(scanb_raw_train.index.equals(scanb_tpm_train.index))
print(scanb_raw_train.index.equals(scanb_raw_subtype_train.index))
print(scanb_raw_test.index.equals(scanb_tpm_test.index))
print(scanb_raw_test.index.equals(scanb_raw_subtype_test.index))
```

```python
# check whether the subtype order matches between raw and tpm training and test sets
print(scanb_raw_subtype_train['subtype'].tolist() == scanb_tpm_subtype_train['subtype'].tolist())
print(scanb_raw_subtype_test['subtype'].tolist() == scanb_tpm_subtype_test['subtype'].tolist())
```

```python
print(scanb_raw_train.shape)
print(scanb_tpm_train.shape)
print(scanb_raw_subtype_train.shape)
print(scanb_raw_test.shape)
print(scanb_tpm_test.shape)
print(scanb_raw_subtype_test.shape)
```

```python
# check subtype proportion in training and test sets
print('Training set:')
print(scanb_raw_subtype_train['subtype'].value_counts())

print('\nTest set:')
print(scanb_raw_subtype_test['subtype'].value_counts())
```

```python
# merge counts and subtype labels for raw count training and test sets
scanb_raw_label_train = pd.merge(scanb_raw_train, scanb_raw_subtype_train, left_index=True, right_index=True)
scanb_raw_label_test = pd.merge(scanb_raw_test, scanb_raw_subtype_test, left_index=True, right_index=True)

# merge counts and subtype labels for tpm count training and test sets
# note: subtype labels are same for both raw and tpm counts
scanb_tpm_label_train = pd.merge(scanb_tpm_train, scanb_raw_subtype_train, left_index=True, right_index=True)
scanb_tpm_label_test = pd.merge(scanb_tpm_test, scanb_raw_subtype_test, left_index=True, right_index=True)
```

```python
# add dataset column in SCAN-B GSE202203 training and test sets for both raw and tpm counts
scanb_raw_label_train['dataset'] = 'SCAN-B Illumina Hiseq'
scanb_tpm_label_train['dataset'] = 'SCAN-B Illumina Hiseq'
scanb_raw_label_test['dataset'] = 'SCAN-B Illumina Hiseq'
scanb_tpm_label_test['dataset'] = 'SCAN-B Illumina Hiseq'
```

```python
# subtype proportion percentage in scanb hiseq training set
scanb_train_classprop = scanb_raw_label_train['subtype'].value_counts()
scanb_train_class = scanb_train_classprop.index.tolist()
scanb_train_classCount = scanb_train_classprop.values.tolist()

for i in range(len(scanb_train_class)):
    percent_train = round((scanb_train_classCount[i]/scanb_raw_label_train.shape[0])*100,2)
    print(f"{scanb_train_class[i]}: {scanb_train_classCount[i]} ({percent_train}%)")
```

```python
# subtype proportion percentage in scanb hiseq test set
scanb_test_classprop = scanb_raw_label_test['subtype'].value_counts()
scanb_test_class = scanb_test_classprop.index.tolist()
scanb_test_classCount = scanb_test_classprop.values.tolist()

for i in range(len(scanb_test_class)):
    percent_test = round((scanb_test_classCount[i]/scanb_raw_label_test.shape[0])*100,2)
    print(f"{scanb_test_class[i]}: {scanb_test_classCount[i]} ({percent_test}%)")
```
