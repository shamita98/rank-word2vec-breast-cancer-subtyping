### Overview

Cleaning and organization of SCAN-B HiSeq and NSCAN-B NextSeq TPM-normalized counts dataset

```python
import pandas as pd
import numpy as np
```

### 1. Prepare the SCAN-B tpm count matrix
Clean the tpm count matrix, such that the column names are genes and row names are samples.

```python
# import raw counts
tpm_counts = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/datasets/SCANB_GSE202203/scanb_uncleaned_data/GSE202203_TPM_Raw_gene_3207.tsv", sep='\t')
```

```python
# view the first three rows 
tpm_counts.head(3)
```

```python
# transpose the raw counts (genes = row names --> column names, samples = column names --> row names)
tpm_counts = tpm_counts.transpose()
```

```python
# view the first three rows
tpm_counts.head(3)
```

```python
# make row 0 as the column names 
tpm_counts.columns = tpm_counts.iloc[0]

# exclude row 0 (duplicate of the column names)
tpm_counts = tpm_counts.iloc[1:,:]
```

```python
# view the first three rows
tpm_counts.head(3)
```

```python
# check the dimensions
tpm_counts.shape
```

```python
# check the sum of all the rows in the raw counts
tpm_counts.sum(axis=1)
```

### 2. Keep PAM50 genes in the raw count matrix

```python
# import PAM50 genes
pam50 = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/datasets/pam50gene_list_csv.csv", header=0)
pam50
```

```python
# check whether PAM50 genes are in the raw counts
tpm_counts.loc[:, tpm_counts.columns.isin(pam50['new_gene_symbol'].tolist())].shape
```

```python
# keep only PAM50 genes in the raw counts
tpm_counts_pam50 = tpm_counts.loc[:, tpm_counts.columns.isin(pam50['new_gene_symbol'].tolist())]
tpm_counts_pam50.shape
```

```python
# check the first three rows
tpm_counts_pam50.head(3)
```

```python
# check whether the gene order is the same
tpm_counts_pam50.columns.tolist() == pam50['new_gene_symbol'].tolist()
```

```python
# rearrange the columns
tpm_counts_pam50 = tpm_counts_pam50[pam50['new_gene_symbol'].tolist()]
tpm_counts_pam50.shape
```

```python
# check whether the gene order is the same after rearranging the columns
tpm_counts_pam50.columns.tolist() == pam50['new_gene_symbol'].tolist()
```

```python
# rename the columns to be ensemble gene ids
tpm_counts_pam50.columns = pam50['ensembl_gene_id'].tolist()
tpm_counts_pam50.head(3)
```

### 3. Extract SCAN-B HiSeq samples and their subtype labels

```python
# import metadata of scan-b hiseq samples annd transpose it
metadata_hiseq = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/datasets/SCANB_GSE202203/scanb_uncleaned_data/GSE202203_GPL11154_HiSeq_series_matrix_metadata.txt", skiprows=38, sep='\t').transpose()
```

```python
# check the dimensions 
metadata_hiseq.shape
```

```python
# check the first three rows
metadata_hiseq.head(3)
```

```python
# make row 0 as the column name
metadata_hiseq_copy = metadata_hiseq.copy()
metadata_hiseq_copy.columns = metadata_hiseq_copy.iloc[0]

# exclude row 0 (duplicate of the column names)
metadata_hiseq_copy = metadata_hiseq_copy.iloc[1:,:]
metadata_hiseq_copy.head(3)
```

```python
# check the dimensions
metadata_hiseq_copy.shape
```

```python
# extract the subtype column
hiseq_subtype = metadata_hiseq_copy.iloc[:,[18]]
hiseq_subtype.columns = ['subtype']
hiseq_subtype.head(3)
```

```python
# check the subtype proportion
hiseq_subtype.value_counts()
```

```python
# subset hiseq samples from the raw counts
tpm_counts_pam50_hiseq = tpm_counts_pam50.loc[tpm_counts_pam50.index.isin(hiseq_subtype.index),:]
tpm_counts_pam50_hiseq.shape
```

```python
# check if the indices match
tpm_counts_pam50_hiseq.index.equals(hiseq_subtype.index)
```

```python
# join raw counts of the samples and their corresponding subtype label
pam50_tpm_counts_subtype_hiseq = tpm_counts_pam50_hiseq.join(hiseq_subtype)
pam50_tpm_counts_subtype_hiseq.shape
```

```python
# remove samples (rows) that are unclassified and normal like
pam50_tpm_counts_subtype_hiseq = pam50_tpm_counts_subtype_hiseq.loc[(pam50_tpm_counts_subtype_hiseq['subtype'] != 'pam50 subtype: Normal') &
(pam50_tpm_counts_subtype_hiseq['subtype'] != 'pam50 subtype: Unclassified'),:]

# check the dimensions of the filtered dataset after removing these samples
print(pam50_tpm_counts_subtype_hiseq.shape)

# check subtype proportion after removing unclassified and normal-like samples
pam50_tpm_counts_subtype_hiseq['subtype'].value_counts()
```

```python
# relabel the classes
label_dict = {'pam50 subtype: LumA': 'LumA', 'pam50 subtype: LumB': 'LumB', 'pam50 subtype: Basal': 'Basal', 'pam50 subtype: HER2': 'Her2'}
pam50_tpm_counts_subtype_hiseq['subtype'] = pam50_tpm_counts_subtype_hiseq['subtype'].replace(label_dict)
pam50_tpm_counts_subtype_hiseq['subtype'].value_counts()
```

```python
# check the first three rows
pam50_tpm_counts_subtype_hiseq.head(3)
```
