### Overview

Extraction of survival data of SCAN-B NextSeq samples for survival analysis

```python
import pandas as pd
import numpy as np
```

### 1. Extract survival data of SCAN-B NextSeq samples

```python
# import clinical metadata of scan-b nextseq samples
metadata_nextseq = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/datasets/SCANB_GSE202203/scanb_uncleaned_data/GSE202203_GPL18573_Nextseq_series_matrix_metadata.txt", skiprows=38, sep='\t').transpose()
```

```python
metadata_nextseq.head(3)
```

```python
# make row 0 as the column name
metadata_nextseq_copy = metadata_nextseq.copy()
metadata_nextseq_copy.columns = metadata_nextseq_copy.iloc[0]

# exclude row 0 (duplicate of the column name)
metadata_nextseq_copy = metadata_nextseq_copy.iloc[1:,:]
metadata_nextseq_copy.head(3)
```

```python
# extract patient characteristics columns
nextseq_patient_char = metadata_nextseq_copy.loc[:, ['!Sample_characteristics_ch1']]
nextseq_patient_char.head(3)
```

```python
# extract survival days, years and events that are in columns 15, 16 and 17
nextseq_survival = nextseq_patient_char.iloc[:,[15, 16, 17]]
nextseq_survival.head(3)
```

```python
# rename column names
nextseq_survival.columns = ['overall_survival_days', 'overall_survival_years', 'overall_survival_event']
nextseq_survival.head(3)
```

### 2. Convert survival data into suitable format

```python
# function to keep only numbers in the survival data
def keep_numbers_in_survival(df, col_names):
    df_copy = df.copy()
    df_nrow = df_copy.shape[0]
    for col in col_names:
        for i in range(df_nrow):
            value = df_copy.loc[df_copy.index[i], col]
            df_copy.loc[df_copy.index[i], col] = value.split(': ')[1]

    return df_copy
```

```python
# keep only the numeric data
nextseq_survival_modified = keep_numbers_in_survival(nextseq_survival, nextseq_survival.columns.tolist())
nextseq_survival_modified.head(3)
```

```python
nextseq_survival_modified.shape
```

```python
# convert data types
nextseq_survival_modified['overall_survival_event'] = pd.to_numeric(nextseq_survival_modified['overall_survival_event'], errors ='coerce').astype('Int64')
nextseq_survival_modified['overall_survival_days'] = pd.to_numeric(nextseq_survival_modified['overall_survival_days'], errors ='coerce').astype('Int64')
nextseq_survival_modified['overall_survival_years'] = pd.to_numeric(nextseq_survival_modified['overall_survival_years'], errors ='coerce').astype('Float64')
```

```python
nextseq_survival_modified.head(3)
```

```python
# check data types
nextseq_survival_modified.dtypes
```

```python
# check the proportion of survival events
nextseq_survival_modified['overall_survival_event'].value_counts()
```

```python
# check for missing values
nextseq_survival_modified.isna().sum(axis=0)
```

```python
# round the overall_survival_years to 3 decimal places
nextseq_survival_modified.loc[:,'overall_survival_years'] = round(nextseq_survival_modified['overall_survival_years'],3)
nextseq_survival_modified.head(3)
```

```python
# get maximum overall survival years in the whole dataset (including normal-like and unclassified samples)
nextseq_survival_modified.loc[nextseq_survival_modified['overall_survival_years'].idxmax(),'overall_survival_years']
```

### 2. Extract survival data of selected SCAN-B NextSeq samples 

These samples include only the four major breast cancer subtypes (Basal, HER2, LumA, LumB)

```python
# import scanb nextseq dataset
scanb_nextseq = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/datasets/SCANB_GSE202203/scanb_cleaned_data/pam50genes_raw_counts_subtype_scanb_nextseq.csv", 
                          header=0, index_col=0)
```

```python
print(scanb_nextseq.shape)
```

```python
# extract only selected nextseq samples (excluding normal-like and unclassified) from the survival dataset
nextseq_selected_survival = nextseq_survival_modified.loc[nextseq_survival_modified.index.isin(scanb_nextseq.index),:]
print(nextseq_selected_survival.shape)
```

```python
print(nextseq_selected_survival['overall_survival_event'].value_counts())
```

```python
# get maximum overall survival years
nextseq_selected_survival.loc[nextseq_selected_survival['overall_survival_years'].idxmax(),'overall_survival_years']
```

```python
# select only the rows from gene expression_subtype dataset that are in survival dataset
# reorder the rows in the gene expression_subtype dataset according to the survival dataset
scanb_nextseq_sorted = scanb_nextseq.loc[nextseq_selected_survival.index,:]
print(scanb_nextseq_sorted.shape)
```

```python
print(scanb_nextseq_sorted.index.equals(nextseq_selected_survival.index))
```

### 3. Merge survival data with subtype labels

```python
# merge survival data of selected nextseq samples with subtype labels
nextseq_survival_subtype = nextseq_selected_survival.join(scanb_nextseq_sorted.loc[:,'subtype'])
print(nextseq_survival_subtype.shape)
```
