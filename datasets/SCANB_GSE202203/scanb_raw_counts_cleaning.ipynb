### Overview

Cleaning and organization of SCAN-B HiSeq and NSCAN-B NextSeq raw counts dataset

```python
import pandas as pd
import numpy as np
```

### 1. Prepare the SCAN-B raw count matrix
Clean the raw count matrix, such that the column names are genes and row names are samples.

```python
# import raw counts
raw_counts = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/datasets/SCANB_GSE202203/scanb_uncleaned_data/GSE202203_RawCounts_gene_3207.tsv", sep='\t')
```

```python
# view the first three rows 
raw_counts.head(3)
```

```python
# transpose the raw counts (genes = row names --> column names, samples = column names --> row names)
raw_counts = raw_counts.transpose()
```

```python
# view the first three rows
raw_counts.head(3)
```

```python
# make row 0 as the column names 
raw_counts.columns = raw_counts.iloc[0]

# exclude row 0 (duplicate of the column names)
raw_counts = raw_counts.iloc[1:,:]
```

```python
# view the first three rows
raw_counts.head(3)
```

```python
# check the dimensions
raw_counts.shape
```

```python
# check the sum of all the rows in the raw counts
raw_counts.sum(axis=1)
```

### 2. Keep PAM50 genes in the raw count matrix

```python
# import PAM50 genes
pam50 = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/datasets/pam50gene_list_csv.csv", header=0)
pam50
```

```python
# check whether PAM50 genes are in the raw counts
raw_counts.loc[:, raw_counts.columns.isin(pam50['new_gene_symbol'].tolist())].shape
```

```python
# keep only PAM50 genes in the raw counts
raw_counts_pam50 = raw_counts.loc[:, raw_counts.columns.isin(pam50['new_gene_symbol'].tolist())]
raw_counts_pam50.shape
```

```python
# check the first three rows
raw_counts_pam50.head(3)
```

```python
# check whether the gene order is the same
raw_counts_pam50.columns.tolist() == pam50['new_gene_symbol'].tolist()
```

```python
# rearrange the columns
raw_counts_pam50 = raw_counts_pam50[pam50['new_gene_symbol'].tolist()]
raw_counts_pam50.shape
```

```python
raw_counts_pam50.head(3)
```

```python
# check whether the gene order is the same after rearranging the columns
raw_counts_pam50.columns.tolist() == pam50['new_gene_symbol'].tolist()
```

```python
# rename the columns to be ensemble gene ids
raw_counts_pam50.columns = pam50['ensembl_gene_id'].tolist()
raw_counts_pam50.head(3)
```

```python
# convert raw data to integers because deseq2 normalization accepts raw counts as inputs and TCGA has integers too
raw_counts_pam50 = raw_counts_pam50.round().astype('int')
```

```python
# check the first three rows
raw_counts_pam50.head(3)
```

### 3. Extract SCAN-B HiSeq samples and their subtype labels

```python
# import metadata of scan-b hiseq samples annd transpose it
metadata_hiseq = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/datasets/SCANB_GSE202203/scanb_uncleaned_data/GSE202203_GPL11154_HiSeq_series_matrix_metadata.txt", skiprows=38, sep='\t').transpose()
```

```python
# check the dimensions 
metadata_hiseq.shape
```

```python
# check the first three rows
metadata_hiseq.head(3)
```

```python
# make row 0 as the column name
metadata_hiseq_copy = metadata_hiseq.copy()
metadata_hiseq_copy.columns = metadata_hiseq_copy.iloc[0]

# exclude row 0 (duplicate of the column names)
metadata_hiseq_copy = metadata_hiseq_copy.iloc[1:,:]
metadata_hiseq_copy.head(3)
```

```python
# check the dimensions
metadata_hiseq_copy.shape
```

```python
# extract the subtype column
hiseq_subtype = metadata_hiseq_copy.iloc[:,[18]]
hiseq_subtype.columns = ['subtype']
hiseq_subtype.head(3)
```

```python
# check the subtype proportion
hiseq_subtype.value_counts()
```

```python
# subset hiseq samples from the raw counts
raw_counts_pam50_hiseq = raw_counts_pam50.loc[raw_counts_pam50.index.isin(hiseq_subtype.index),:]
raw_counts_pam50_hiseq.shape
```

```python
# check if the indices match
raw_counts_pam50_hiseq.index.equals(hiseq_subtype.index)
```

```python
# check for low-quality samples (characterized by having lots of low-count genes)
# check for rows with less than 10 raw read counts in all the gene columns
# 10 is the threshold for low count genes according to DESeq2 analysis documentation
print((raw_counts_pam50_hiseq.iloc[:, 0:50] <= 10).all(axis=1).sum())

# check rows with less than 10 raw read counts in at least half of the gene columns
print(((raw_counts_pam50_hiseq.iloc[:, 0:50] <= 10).sum(axis=1) >= raw_counts_pam50_hiseq.iloc[:, 0:50].shape[1] // 2).sum())
```

```python
# join raw counts of the samples and their corresponding subtype label
pam50_raw_counts_subtype_hiseq = raw_counts_pam50_hiseq.join(hiseq_subtype)
pam50_raw_counts_subtype_hiseq.shape
```

```python
# remove samples (rows) that are unclassified and normal like
pam50_raw_counts_subtype_hiseq = pam50_raw_counts_subtype_hiseq.loc[(pam50_raw_counts_subtype_hiseq['subtype'] != 'pam50 subtype: Normal') &
(pam50_raw_counts_subtype_hiseq['subtype'] != 'pam50 subtype: Unclassified'),:]

# check the dimensions of the filtered dataset after removing these samples
print(pam50_raw_counts_subtype_hiseq.shape)

# check subtype proportion after removing unclassified and normal-like samples
pam50_raw_counts_subtype_hiseq['subtype'].value_counts()
```

```python
# relabel the classes
label_dict = {'pam50 subtype: LumA': 'LumA', 'pam50 subtype: LumB': 'LumB', 'pam50 subtype: Basal': 'Basal', 'pam50 subtype: HER2': 'Her2'}
pam50_raw_counts_subtype_hiseq['subtype'] = pam50_raw_counts_subtype_hiseq['subtype'].replace(label_dict)
pam50_raw_counts_subtype_hiseq['subtype'].value_counts()
```

```python
# check the first three rows
pam50_raw_counts_subtype_hiseq.head(3)
```
