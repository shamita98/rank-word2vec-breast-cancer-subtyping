```python
# import go bp terms of each cluster
cluster_1 = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/pam50_classification_word2vec/gene_ontology_word2vec_gene_clusters/word2vec_gene_clusters_go_bp/cluster_1_go_bp.csv",
                       index_col=0)
cluster_2 = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/pam50_classification_word2vec/gene_ontology_word2vec_gene_clusters/word2vec_gene_clusters_go_bp/cluster_2_go_bp.csv",
                       index_col=0)
cluster_3 = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/pam50_classification_word2vec/gene_ontology_word2vec_gene_clusters/word2vec_gene_clusters_go_bp/cluster_3_go_bp.csv",
                       index_col=0)
cluster_4 = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/pam50_classification_word2vec/gene_ontology_word2vec_gene_clusters/word2vec_gene_clusters_go_bp/cluster_4_go_bp.csv",
                       index_col=0)
cluster_5 = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/pam50_classification_word2vec/gene_ontology_word2vec_gene_clusters/word2vec_gene_clusters_go_bp/cluster_5_go_bp.csv",
                       index_col=0)
cluster_6 = pd.read_csv("C:/Users/User/Documents/master_thesis_project_analysis/pam50_classification_word2vec/gene_ontology_word2vec_gene_clusters/word2vec_gene_clusters_go_bp/cluster_6_go_bp.csv",
                       index_col=0)
```

```python
## write a function to do the following:
# remove duplicates based on geneID column and keep only the first occurrence
# sort based on Count (which is number of genes covered)
# remove rows where Count = 1 (where only one gene is covered)
# return the modified GO BP term df

def clean_go_bp(go_bp_df):
    go_bp_df_unique = go_bp_df.sort_values(by=['p.adjust'], ascending=True)
    go_bp_df_unique = go_bp_df.drop_duplicates(subset=['geneID'], keep='first')
    go_bp_df_unique = go_bp_df_unique.sort_values(by=['Count'], ascending=False)
    go_bp_df_unique = go_bp_df_unique.loc[go_bp_df_unique['Count'] != 1,:]

    return go_bp_df_unique
```

```python
## write a function to do the following:
# select the first geneID (longest or highest coverage) as the first reference in a ref array
# if the other geneID entries are subsets of the ref, ignore it
# if the other geneID entries are not subsets of the ref, add it as another ref
# store the indices of all the refs in an array
# keep only the indices of the ref in the df and return it

def keep_unique_superset_go_bp(go_bp_df):
    ref_gene_ids = []
    ids_to_keep = []

    for id in go_bp_df.index:
        gene_id = go_bp_df.loc[go_bp_df.index == id, ['geneID']].values[0][0]
        gene_id_list = gene_id.split('/')
    
        if len(ref_gene_ids) == 0:
            ref_gene_ids.append(gene_id_list)
            ids_to_keep.append(id)
        else:
            match = False
            for ref in ref_gene_ids:
                if set(gene_id_list).issubset(set(ref)):
                    match = True
                    break
                    
            if not match:
                ref_gene_ids.append(gene_id_list)
                ids_to_keep.append(id)

    go_bp_df_ref = go_bp_df.loc[ids_to_keep, :]
    return(go_bp_df_ref)
```

```python
# clean the go bp terms of each gene cluster
cluster_1_unique = clean_go_bp(cluster_1)
cluster_2_unique = clean_go_bp(cluster_2)
cluster_3_unique = clean_go_bp(cluster_3)
cluster_4_unique = clean_go_bp(cluster_4)
cluster_5_unique = clean_go_bp(cluster_5)
cluster_6_unique = clean_go_bp(cluster_6)

print(cluster_1_unique.shape)
print(cluster_2_unique.shape)
print(cluster_3_unique.shape)
print(cluster_4_unique.shape)
print(cluster_5_unique.shape)
print(cluster_6_unique.shape)
```

```python
# remove geneID entries that are subsets, keep unique supersets
cluster_1_superset = keep_unique_superset_go_bp(cluster_1_unique)
cluster_2_superset = keep_unique_superset_go_bp(cluster_2_unique)
cluster_3_superset = keep_unique_superset_go_bp(cluster_3_unique)
cluster_4_superset = keep_unique_superset_go_bp(cluster_4_unique)
cluster_5_superset = keep_unique_superset_go_bp(cluster_5_unique)
cluster_6_superset = keep_unique_superset_go_bp(cluster_6_unique)

print(cluster_1_superset.shape)
print(cluster_2_superset.shape)
print(cluster_3_superset.shape)
print(cluster_4_superset.shape)
print(cluster_5_superset.shape)
print(cluster_6_superset.shape)
```
